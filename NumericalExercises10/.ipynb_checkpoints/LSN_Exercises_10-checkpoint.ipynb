### Exercise 10.1

By adapting your Genetic Algorithm (GA) code, developed during the Numerical Exercise 9, write a C++ code to solve the TSP with a **Simulated Annealing** (SA) algorithm. Apply your code to the optimization of a path among

- a picture of the length of the best path as a function of the iteration of your algorithm
- a picture of the best path

and compare your SA results with those obtained with GA in the previous numerical exercise.

### Solution

The simulated annealing (SA) algorithm requires an annealing schedule made of $N$ (number of cooling steps) increasing values of inverse temperature $\beta_i$ and integer numbers $n_i$, $i=1,...,N$. Starting from an initial randomly generated path (the same class of exercise 09 is used), each cooling step consists in performing $n_i$ Metropolis moves that are accepted according to the Boltzmann-like (non-normalized) probability $e^{-\beta_i E}$, where the energy $E$ is represented by the cost function of the optimization problem, for which the path length $L^{(1)}$ is used. The acceptance probability is then

$$
\alpha = min \left( 1, e^{-\beta_i \left(L^{(1)}(x') - L^{(1)}(x) \right)} \right)
$$

The move from the old path _x_ to a new path *x'* is suggested by applying the swap, shift and inversion mutation operators, whose probability are increased with respect to the GA to $40\%, 30\%, 30\%$ respectively, because the crossover is absent here. The temperature is then reduced ($\beta$ is increased) and another step is taken and so on until the temperature is close to $0$ and at that point the path with minimum energy (i.e. minimum length) should be found.

#### Cities on a circumference

$32$ cities randomly placed on the unitary circumference.

$N=2\times10^4$ cooling steps: $\beta_i=i\times1/100$, $n_i=100$, $i=1,...,N$. These parameters have been chosen after several attempts as they return the best value of the minimum. The starting temperature ($100$) is high enough to explore all the possible configurations without getting trapped into a local minimum, while the final temperature ($1/200$) is low enough to have the best approximation of the minimum. Most of the times once $N$ and $\beta_i$ are fixed, increasing $n_i$ beyond $100$ doesn't really bring to any improvement in the performances. Indeed $100$ steps should be enough for the Metropolis to reach a high-probability region of the specific distribution (depending on $\beta_i$) that we are sampling.

The best path length is $6.2403$, the same value obtained with the GA.

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

length=np.loadtxt("10.1_SA/length_circumference.out")

plt.figure(figsize=(15,8))
plt.plot(length[:,0], length[:,1], linewidth=0.5, color='mediumblue')
plt.grid(True)
plt.title('Evolution of path length, circumference')
plt.xlabel('$beta$')
plt.ylabel('$L^{(1)}$')
plt.show()
```

```python
best_path=np.loadtxt("10.1_SA/best_path_circumference.out")

plt.figure(figsize=(10,10))
plt.plot(best_path[:,0], best_path[:,1], marker='o', color='limegreen', markerfacecolor='darkgreen', markersize=8)
plt.grid(True)
plt.title('Best path, circumference')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

### Exercise 10.2

Parallelize with MPI libraries your Genetic Algorithm code in order to solve the TSP by performing a *Genetic Search* with 4 **parallel GA searches of the optimal path**:
each node should perform an independent GA search the so-called *Continents*, **but** every $N_{migr}$ generations the *Continents* should exchange their best individuals randomly.
Apply your code to the TSP problem for 32 cities randomly placed inside a square and try to estimate if your performances improve with *Continents*.

### Solution

The only difference in the code <font face="Courier">TSP.cpp</font> with respect to that of Numerical Exercise 9, apart from the MPI initialization and finalization, is the function _Migrate_, which is called every $N_{migr}$ iterative steps and exploits a _universal_ random number generator. The difference between this generator and _rnd_ is that the first one is initialized using the same seed on all nodes, while for the second one different seeds are used on each node. _universal_ will then always produce the same sequence, while the _rnd_ sequence is node-dependent. This trick allows all continents to know where all the best individuals have to move without one telling all the others (e.g. with *MPI_Bcast*). _Migrate_ uses _universal_ to randomly create a migration scheme, namely a permutation $\sigma$ of $1,...,N_{nodes}$ where $\sigma_i\neq i$ for all $i$ except $i=N_{nodes}$ (since the $\sigma_i$ are produced in order, if the first $N_{nodes}-1$ happen to be a permutation of $1,...,N_{nodes}-1$, the last one must necessarily be $\sigma_{N_{nodes}}=N_{nodes}$). At this point each continent can send its best individual to the destination continent and receive the one of the origin continent. This is realized by means of the blocking routines *MPI_Send* and *MPI_Recv*.

#### Simulation

The parallelized GA has been applied to the search of the minimum path among 32 cities inside the square with the following parameters:

$N=2\times10^4$ iterations, $N_{migr}=100$

$P_{swap}=20\%$, $P_{shift}=10\%$, $P_{inversion}=10\%$

Using $4$ nodes the best path length returned is $4.63666$, the same as the SA of exercise 10.1 and better than exercise 09.1. The results obtained from the simple and the parallelized GA with the same parameters are compared in the following sections. The minimum length returned by the non-parallelized algorithm in these conditions is $4.89730$.

##### Search of the best path

The graph below compares the lenght $L^{(1)}$ of the best individual and the average length $\langle L^{(1)} \rangle$ of the best half of the population for the continent that returned the shortes path in the parallelized simulation and for the only continent of the non-parallelized simulation. The parallelized algorithm converges faster and to a better minimum path than the non-parallelized one (in the same conditions). We could have expected an imporvement in the performance, since 4 nodes can find a better solution than one and once it is found in one continent it is quickly spread all over the world thanks to the migrations.

Some equally spaced rectangular steps can be noticed. They clearly occur when during a migration the continent receives an indivdual that is worse than his best individual. Nevertheless they quickly disappear as the fitting characteristics of the rest of the population prevail.

```python
length=np.loadtxt("10.2_parallelGA/length.out")
lengthex9=np.loadtxt("10.2_parallelGA/lengthex9.out")
step=range(len(length[:,0]))

plt.figure(figsize=(15,8))
plt.plot(step, lengthex9[:,0], label='best $L^{(1)}$ simple GA', color='mediumblue')
plt.plot(step, lengthex9[:,1], label='$<L^{(1)}>$ simple GA', color='aqua')
plt.plot(step, length[:,0], label='best $L^{(1)}$ parallelized GA', color='firebrick')
plt.plot(step, length[:,1], label='$<L^{(1)}>$ parallelized GA', color='lightsalmon')
plt.grid(True)
plt.title('Path length, square')
plt.xlabel('iterations')
plt.ylabel('$L^{(1)}$')
plt.legend()
plt.show()
```

##### Best path

```python
best_path=np.loadtxt("10.2_parallelGA/best_path.out")
best_pathex9=np.loadtxt("10.2_parallelGA/best_pathex9.out")

plt.figure(figsize=(10,10))
plt.plot(best_path[:,0], best_path[:,1], marker='o', color='gold', markerfacecolor='darkgreen', markersize=8, label='parallelized GA: $L^{(1)}=4.63666$')
plt.plot(best_pathex9[:,0], best_pathex9[:,1], marker='o', color='limegreen', markerfacecolor='darkgreen', markersize=8, ls=':', label='simple GA: $L^{(1)}=4.89730$')
plt.grid(True)
plt.title('Best path, square')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend(loc='best')
plt.show()
```
