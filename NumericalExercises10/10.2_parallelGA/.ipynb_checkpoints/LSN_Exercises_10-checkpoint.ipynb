### Exercise 10.1

By adapting your Genetic Algorithm (GA) code, developed during the Numerical Exercise 9, write a C++ code to solve the TSP with a **Simulated Annealing** (SA) algorithm. Apply your code to the optimization of a path among 

- 32 cities randomly placed on a circumference

Show your results via:

- a picture of the length of the best path as a function of the iteration of your algorithm
- a picture of the best path

and compare your SA results with those obtained with GA in the previous numerical exercise.

### Results

SA was performed with increasing values of $\beta$ (inverse of the temperature), which were multiples of $1/300$. At each temperature the same number of Metropolis steps were executed.

Each step consists in operating a swap, a shift and an inversion mutation on path __*x*__, with probabilities $40\%$, $30\%$ and $ 30\%$ respectively and then accepting the new path **_x'_** with probability $P=min(1,\ e^{-\beta(L(x')-L(x))})$, where $L(x)=L^{(1)}(x)$ is the length of the path.

#### Cities on a circumference

$32$ cities randomly placed on the unitary circumference.

$20000$ cooling steps (values of $\beta$), each with $100$ Metropolis steps.

The best path's length is $6.2403$, the same value obtained with GA.

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

def readFile(filename):
    with open(filename, "r") as inp:
        read = inp.readlines()
        data=[line.split() for line in read]
        Tdata = list(zip(*data))
        for i, _ in enumerate(Tdata):
            Tdata[i]=list(Tdata[i])
            for j, _ in enumerate(Tdata[i]):
                Tdata[i][j]=float(Tdata[i][j])
    return Tdata
```

```python
length=readFile("length_circumference.out")

plt.figure(figsize=(8,6))
plt.plot(length[0], length[1])
plt.grid(True, which='both')
plt.title('path length')
plt.xlabel('$beta$')
plt.ylabel('$L^{(1)}$')
plt.show()
```

```python
best_path=readFile("best_path_circumference.out")

plt.figure(figsize=(6,6))
plt.plot(best_path[0], best_path[1])
plt.grid(True)
plt.title('best path')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
