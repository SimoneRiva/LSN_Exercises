In Statistical Mechanics the one-dimensional (1D) Ising Model is exactly resoluble. The Hamiltonian for a system of N spins with nearest neighbors interaction is given by (in the following I will assume units such that $\mu_B=1$ and $k_B=1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

where $\{ s_i \}_{i=1,N}$ are spin variables taking values in $\{-1;1\}$, $J>0$ (ferromagnetic model),and with the assumption $s_{N+1} = s_{1}$.

The partition function for $N$ spins, $Z$, can be exactly computed and it turns out to be:

$$
Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N
$$

where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_b T} = \frac{1}{T}$.
Once the partition function has been computed, the Helmholtz free energy is readly obtained as $A = -k_B T \ln Z$. All the thermodynamic quantities of interest can be deduced from $A$ or, equivalently, from $Z$:

- the **Internal Energy** per spin (with $h=0$):

$$
\frac{U(N,T)}{N} = \frac{\langle H \rangle}{N}
= \frac{1}{N} \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}
= -\frac{1}{N}\frac{\partial \ln Z}{\partial \beta}
\underset{h=0}= -J \tanh (\beta J) \Biggl( \frac{1 + \tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

- the **Heat Capacity** (with $h=0$):

$$
C(N,T) = \frac{\partial U(N,T)}{\partial T}
= \frac{\partial \beta}{\partial T}\frac{\partial U}{\partial \beta}
= -k_B\beta^2 \frac{\partial}{\partial \beta} \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z} \right] = \\
= -k_B\beta^2 \left[ -\sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \sum_{\{ s_i \}} H \exp \left( -\beta H\right)\frac{\partial Z^{-1}}{\partial \beta} \right] = \\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} + \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]\frac{1}{Z}\frac{\partial Z}{\partial \beta} \right\} =\\
= k_B\beta^2 \left\{ \sum_{\{ s_i \}} H^2 \frac{\exp \left( -\beta H\right)}{Z} - \left[ \sum_{\{ s_i \}} H \frac{\exp \left( -\beta H\right)}{Z}\right]^2 \right\} =
k_B\beta^2 \left(\langle H^2 \rangle -\langle H \rangle^2 \right)
$$

$$
\frac{C(N,T)}{N} = \frac{1}{N}\frac{\partial U(N,T)}{\partial T} =\\
\underset{h=0}= 
k_B (\beta J)^2 \left\{ 1+\tanh^N (\beta J)+(N-1)\tanh^2 (\beta J)+(N-1)\frac{\tanh^{N-2} (\beta J)}{1+\tanh^N (\beta J)}
- N \left[ \frac{\tanh (\beta J) + \tanh^{N-1} (\beta J)}{1+\tanh^N (\beta J)} \right]^2 \right\}
$$

- The **magnetization** (with $h \not= 0$):

Note that with $h=0$ the magnetization $M(N,T,h=0)=\langle \sum_{i=1,N}s_i \rangle =0$ because in 1D we cannot have a (ferromagnetic) phase transition in absence of an external magnetic field.

$$
M(N,T,h) = -\frac{\partial A}{\partial h} = k_B T \frac{\partial \ln Z}{\partial h} =
\frac{1}{\beta} \frac{1}{Z}\frac{\partial}{\partial h}\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} = \left\langle \sum_{i=1,N}s_i \right\rangle
$$

The exact formula from $M(N,T,h) = k_B T \dfrac{\partial \ln Z}{\partial h}$ is:
$$
\frac{M(N,T,h)}{N}=\frac{e^{\beta J}\; \sinh(\beta h)}{Z} \; \Bigl[\lambda_1^{N-1}\cdot \Bigl(1+\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr) + \lambda_2^{N-1}\cdot \Bigl(1-\frac{e^{\beta J}\; \cosh(\beta h)}{\sqrt{e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J)}}\Bigr)\Bigr]
$$

- The **magnetic susceptibility** (with $h=0$):

$$
\chi(N,T) = \frac{\partial M}{\partial h} = 
\frac{\partial}{\partial h}\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z} =\\
= \beta\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right]^2 \frac{\exp \left( -\beta H\right)}{Z}
-\sum_{\{ s_i \}} \left[ \sum_{i=1,N}s_i\right] \frac{\exp \left( -\beta H\right)}{Z}
\frac{1}{Z}\frac{\partial Z}{\partial h} = 
\beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle - \left\langle \sum_{i=1,N}s_i \right\rangle^2 \right] = \\
\underset{h=0}= \beta \left[ \left\langle \left( \sum_{i=1,N}s_i \right)^2 \right\rangle \right] = \beta e^{2\beta J} \Biggl( \frac{1-\tanh^N (\beta J)}{1+\tanh^N (\beta J)} \Biggl)
$$

### Solution

The code <font face="Courier">Monte_Carlo_ISING_1D.cpp</font> has been modified to read an input file <font face="Courier">input.dat</font> containing a restart boolean-like integer variable in the first line. If this variable is set to $1$ the program will run a simulation starting from the configuration read from <font face="Courier">config.0</font>, otherwise it will generate the starting configuration randomly.

Other changes include the function _Boltzmann_, which is modified to evaluate the total energy variation when flipping one spin taking the current spin value, instead of the spin variation. The calculation of the total energy variation when flipping $s_i$ is straightforward:

$$
\Delta H = -J \Delta s_i (s_{i-1}+s_{i+1}) - h \Delta s_i = 2J s_i (s_{i-1}+s_{i+1}) + 2h s_i
$$

since the variation of the spin is obviously $\Delta s_i = -2 s_i$, where $s_i$ is its value before getting flipped.

When sampling with the Metropolis algorithm, after randomly drawing one spin its flipping is accepted with probability $\alpha=min(1, e^{-\beta \Delta H})$. When sampling with the Gibbs algorithm the chosen spin is set to $1$ with probability $\alpha = \left( 1+e^{-\beta [2J (s_{i-1}+s_{i+1})+2h]} \right) ^{-1}$, using the fact that the argument of the exponential (apart from the factor $-\beta$) coincides with the output of the function _Boltzmann_ with current spin $1$ (see eq. above), otherwise it is set to $-1$.

The calculation of $C(N,T)$, $\chi(N,T)$ and $M(N,T,h=0.02)$ has been implementented in the functions _Measure_ and _Averages_. The function _Accumulate_ accumulates the total energy and magnetization of all spin configurations within a block and their squares. This allows _Averages_ to compute $\langle H \rangle$, $\langle H^2 \rangle$, $\langle \sum_{i=1}^N s_i \rangle$ and $\langle \left( \sum_{i=1}^N s_i \right) ^2 \rangle$ and therefore the blocking averages and errors of all the 4 quantities we are intersted in.

#### Simulations

Simulations of $M=2\times10^5$ MC steps divided into $20$ blocks have been performed using both the Metropolis and the Gibbs algorithm, restarting from the final configuration of a previous $10^4$ steps equilibration simulation. The parameters used are: $N_{spin}=50$, $J=1$, $h=0$ (except for the measure of magnetization, where $h=0.02$) and $T$ varying from $0.5$ to $2.0$ by steps of $0.1$.

The final averages and uncertainties for $U/N(T)$, $C(T)$, $\chi(T)$ and $M/N(T,h=0.02)$ computed with all blocks are plotted in the following pictures as functions of the temperature. All plots from both the Metropolis and the Gibbs sampling display a good compatibility with the exact curves obtained by analytical calculations.

#### Internal energy per spin

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

points=100
T = np.linspace(0.2,3.0,num=points)
t=np.linspace(0.5,2.0,16)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

plt.figure(figsize=(18,14))
plt.plot(T, e, label='theory')
energy=np.loadtxt("data/metro.ene.dat")
plt.errorbar(t, energy[:,0], yerr=energy[:,1], label='Metropolis')
energy=np.loadtxt("data/gibbs.ene.dat")
plt.errorbar(t, energy[:,0], yerr=energy[:,1], label='Gibbs')
plt.title('Ising 1D, internal energy')
plt.xlabel('$T$')
plt.ylabel('$U/N$')
plt.legend()
plt.grid()
plt.show()
```

#### Heat capacity

```python
h=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

plt.figure(figsize=(18,14))
plt.plot(T, h, label='theory')
heat=np.loadtxt("data/metro.heat.dat")
plt.errorbar(t, heat[:,0], yerr=heat[:,1], label='Metropolis')
heat=np.loadtxt("data/gibbs.heat.dat")
plt.errorbar(t, heat[:,0], yerr=heat[:,1], label='Gibbs')
plt.title('Ising 1D, heat capacity')
plt.xlabel('$T$')
plt.ylabel('$C$')
plt.legend()
plt.grid()
plt.show()
```

#### Magnetization per spin

```python
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

plt.figure(figsize=(18,14))
plt.plot(T, M, label='theory')
mag=np.loadtxt("data/metro.mag.dat")
plt.errorbar(t, mag[:,0], yerr=mag[:,1], label='Metropolis')
mag=np.loadtxt("data/gibbs.mag.dat")
plt.errorbar(t, mag[:,0], yerr=mag[:,1], label='Gibbs')
plt.title('Ising 1D, magnetization with h = 0.02')
plt.xlabel('$T$')
plt.ylabel('$M/N$')
plt.legend()
plt.grid()
plt.show()
```

#### Magnetic susceptibility

```python
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

plt.figure(figsize=(18,14))
plt.plot(T, X, label='theory')
chi=np.loadtxt("data/metro.chi.dat")
plt.errorbar(t, chi[:,0], yerr=chi[:,1], label='Metropolis')
chi=np.loadtxt("data/gibbs.chi.dat")
plt.errorbar(t, chi[:,0], yerr=chi[:,1], label='Gibbs')
plt.title('Ising 1D, magnetic susceptibility')
plt.xlabel('$T$')
plt.ylabel('$\chi$')
plt.legend()
plt.grid()
plt.show()
```
