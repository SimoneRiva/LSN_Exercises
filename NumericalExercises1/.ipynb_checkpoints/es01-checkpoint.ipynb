```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x=np.array([])
sum_prog=np.array([])
err_prog=np.array([])

with open("results1.1.txt", "r") as inp:
    data = inp.readlines()
    for line in data:
        num = line.split()
        x=np.append(x, float(num[0]))
        sum_prog=np.append(sum_prog, float(num[1]))
        err_prog=np.append(err_prog, float(num[2]))


plt.errorbar(x,sum_prog,yerr=err_prog)
plt.xlabel('#throws')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```
