```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x=np.array([])
sumi_prog=np.array([])
erri_prog=np.array([])
sumj_prog=np.array([])
errj_prog=np.array([])

with open("results2.1.out", "r") as inp:
    data = inp.readlines()
    for line in data:
        num = line.split()
        x=np.append(x, float(num[0]))
        sumi_prog=np.append(sumi_prog, float(num[1]))
        erri_prog=np.append(erri_prog, float(num[2]))
        sumj_prog=np.append(sumj_prog, float(num[3]))
        errj_prog=np.append(errj_prog, float(num[4]))


plt.errorbar(x,sumi_prog,yerr=erri_prog)
plt.xlabel('#throws')
plt.ylabel('<I>-1')
plt.grid(True)
plt.show()

plt.errorbar(x,sumj_prog,yerr=errj_prog)
plt.xlabel('#throws')
plt.ylabel('<I>-1')
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import curve_fit

step=np.array([])
ave_d=np.array([])
err_d=np.array([])
ave_c=np.array([])
err_c=np.array([])

with open("results2.2.out", "r") as inp:
    data = inp.readlines()
    for line in data:
        num = line.split()
        step=np.append(step, float(num[0]))
        ave_d=np.append(ave_d, float(num[1]))
        err_d=np.append(err_d, float(num[2]))
        ave_c=np.append(ave_c, float(num[3]))
        err_c=np.append(err_c, float(num[4]))

def func(N, k):
    return k * np.sqrt(N)

popt, pcov = curve_fit(func, step, ave_d)

plt.errorbar(step,ave_d,yerr=err_d, label='data')
plt.plot(step, func(step, *popt), 'r-', label='fit $k\sqrt{N}$: $k=%5.3f$' % tuple(popt))
plt.xlabel('Nsteps')
plt.ylabel('length')
plt.legend(loc='best')
plt.grid(True)
plt.show()

popt, pcov = curve_fit(func, step, ave_c)

plt.errorbar(step,ave_c,yerr=err_c, label='data')
plt.plot(step, func(step, *popt), 'r-', label='fit $k\sqrt{N}$: $k=%5.3f$' % tuple(popt))
plt.xlabel('Nsteps')
plt.ylabel('length')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

As we have seen, from the diffusion equation
\begin{equation}
\frac{\partial p (x,t)}{\partial t} = D \frac{\partial^2 p (x,t)}{\partial x^2},
\end{equation}
with starting condition $p(x,t=0)=\delta(x)$ (i. e. the random walk starts from the origin at $t=0$ with probability 1), one obtains as a solution the Gaussian centered in the origin and with variance $\sigma^2=2Dt$,
\begin{equation}
p (x,t) = \frac{1}{\sqrt{4 \pi Dt}} e^{- \frac{x^2}{4Dt}},
\end{equation}
from which it is straightforward to obtain:
\begin{equation}
<x^2(t)>_{RW}=2Dt.
\end{equation}
Since we are computing the square root of this value it makes sense that, if the behaviour is diffusive, it will grow proportionally to the square root of $N_{steps}$ (which stands for the time in this discrete case). Moreover, since from the fit we have found the value $k=1$, we can deduce that the diffusion coefficient $D=\frac{1}{2}$.
