### Solution

Simulating a system with the Monte Carlo NVT code consists in producing a possible evolution of a statistical canonical ensemble of $N$ particles by exploring its phase space through the sampling of the Boltzmann weight $e^{-\beta U}/Z$. Here $\beta$ is fixed by the choice of the system temperature, while the total potential energy

$$U=U(\vec{r_1},...,\vec{r_N})=\frac{1}{2} \sum_{i,j} V_{LJ} (|\vec{r_i}-\vec{r_j}|)$$

is in this case the sum of pairwise Lennard-Jones interaction energies among particles. In this section all graphs are in Lennard-Jones units.

#### Equilibration

The sampling is realized by means of the Metropolis algorithm starting from a fcc crystalline configuration. Therefore the simulation needs to be first equilibrated, in order to reach a high-probablity region of the phase space and avoid sampling a lot of unlikely states in the first steps. One expects the system to equilibrate in few steps when in the solid phase, since its equilibrium configuration will be similar to a fcc crystal apart from thermal oscillations of the atoms around their equilibrium positions, and the liquid and gas phase to take progressively more steps because of the increasing disorder. To check this the code has been modified to print on files called <font face="Courier">measures/instant_energy.out</font> and <font face="Courier">measures/instant_pressure.out</font> the measured values of potential energy per particle $U/N$ and pressure $P$ after each MC step. These values are plotted in the following pictures for the starting steps of an initial equilibration simulation of $M=5\times 10^4$ total steps. One MC step consists in trying to move a randomly selected particle as many times as the particles in the system.

#### Autocorrelation and statistical uncertainties

A longer simulation of $5\times 10^5$ steps has been run starting from the final configuration of the previous one. The following cells use the instantaneous measures to compute the autocorrelation function and the final uncertainty of the averages as a function of the size of blocks they are divided into for both the potential energy and the pressure.

```python
# Be careful, this cell takes a while to run

energy_solid=np.loadtxt("measures/instant_energy_solid.out")
pressure_solid=np.loadtxt("measures/instant_pressure_solid.out")

energy=energy_solid
pressure=pressure_solid

n_prop=2
N=len(energy)
iv=0
ip=1
tmax=200

def ComputeAutocorrelation():
    ave2=[0, 0]
    ave_2=[0, 0]
    sum=[0, 0]
    sigma2=[0, 0]
    chi=np.ndarray((n_prop, tmax))
    for t in range(tmax):
        for p in range(n_prop):
            sum[p]=0
            ave2[p]=0
            ave_2[p]=0
        for tp in range(N-t):
            ave2[iv]=ave2[iv]+energy[tp]
            ave_2[iv]=ave_2[iv]+energy[tp]*energy[tp]
            ave2[ip]=ave2[ip]+pressure[tp]
            ave_2[ip]=ave_2[ip]+pressure[tp]*pressure[tp]
            sum[iv]=sum[iv]+energy[tp]*energy[tp+t]
            sum[ip]=sum[ip]+pressure[tp]*pressure[tp+t]
        for p in range(n_prop):
            ave2[p]=(ave2[p]/(N-t))*(ave2[p]/(N-t))
            ave_2[p]=ave_2[p]/(N-t)
            sigma2[p]=ave_2[p]-ave2[p]
            sum[p]=sum[p]/(N-t)
            chi[p][t]=(sum[p]-ave2[p])/sigma2[p]
    return chi

chi_solid=ComputeAutocorrelation()
```

```python
import math

def error (sum, sum2, i):
    if(i==0):
        return 0
    else:
        return math.sqrt((sum2-sum*sum)/i)

Ls=[int(x) for x in np.logspace(1, 3+math.log10(5), 30)]
l=len(Ls)

def ComputeErrors():
    ave=np.ndarray((n_prop, l))
    errors=np.ndarray((n_prop, l))
    il=0
    for L in Ls:
        M=int(N/L)
        ave_block=np.zeros((n_prop, M))
        ave_2block=np.zeros((n_prop, M))
        err=np.ndarray((n_prop, M))
        for i in range(M):
            sum=[0,0]
            for j in range(L):
                k=i*L+j
                sum[iv]=sum[iv]+energy[k]
                sum[ip]=sum[ip]+pressure[k]
            for p in range(n_prop):
                sum[p]=sum[p]/L
                ave_block[p][M-1]=ave_block[p][M-1]+sum[p]                #
                ave_2block[p][M-1]=ave_2block[p][M-1]+sum[p]*sum[p]       #

    # If you want to compute the progressive averages and errors
    # for every number of blocks instead than only with all the blocks
    # (i.e. only ave_block[p][M-1] and err[p][M-1]), just remove the '#s'
    # on the left and comment the lines with the '#' on the right.
    
    #            for j in range(M-i):
    #                k=M-j-1
    #                ave_block[p][k]=ave_block[p][k]+sum[p]
    #                ave_2block[p][k]=ave_2block[p][k]+sum[p]*sum[p]
    #            ave_block[p][i]=ave_block[p][i]/(i+1)
    #            ave_2block[p][i]=ave_2block[p][i]/(i+1)
    #            err[p][i]=error(ave_block[p][i], ave_2block[p][i], i)
        for p in range(n_prop):
            ave_block[p][M-1]=ave_block[p][M-1]/M                         #
            ave_2block[p][M-1]=ave_2block[p][M-1]/M                       #
            err[p][M-1]=error(ave_block[p][M-1], ave_2block[p][M-1], M-1) #
            ave[p][il]=ave_block[p][M-1]
            errors[p][il]=err[p][M-1]
        il=il+1
    return errors

energy=energy_solid
pressure=pressure_solid
errors_solid=ComputeErrors()
```

- Solid phase: the autocorrelation functions of both $U/N$ and $P$ become essentially null for $\tau \approx 50$, so blocks should have size $L \gg 50$ to avoid correlations between the average potential energies and pressures of two consecutive blocks.

- Gas phase: the autocorrelation function of $P$ becomes essentially null for $\tau \approx 100$. The autocorrelation of $U/N$ decreases much more slowly, but after $1000$ steps it is small enough to say that if we choose $L=5\times 10^3$ blocks, their data should be weakly correlated.

In the pictures above the statistical uncertainties are plotted as a function of the block size $L$ for 30 values of $L$ logarithmically distributed between $10$ and $5 \times 10^3$. For small blocks the uncertainties increase with $L$, because the smaller the block size is the more correlated two consecutive blocks are, and thus the statistical error is underestimated because of non-independence of the measures. However for large enough $L$, when different blocks produce independent measures, errors saturate to their real statistical value. This happens at $L \approx 5\times10^2,1\times10^3, 3\times10^3$ for the solid, liquid and gas phase respectively, which confirms that $5\times10^3$ is a large enough size to have statistically uncorrelated blocks.

### Solution

The following formula is used to compute $g(r)$:

$$
g(r) = \frac{1}{\rho N \Delta V(r)} \langle \sum_{i\neq j=1}^N \delta \left[ |\vec{r}| - |\vec{r_i}-\vec{r_j}| \right] \rangle
$$

where $\Delta V(r) =\frac{4\pi}{3} \left[ (r+dr)^3-r^3 \right]$.

The update of the histogram of $g(r)$ has been added to the function _Measure_: the interval $[0,L/2]$ is divided in $100$ bins and a bin counter is increased by $2$ every time a couple of particles is found at the corresponding distance $r$ (every couple is counted only once in the loop). _Averages_ then normalizes all the block averages of the bins counts dividing them by $\rho N \Delta V(r)$ and calculates the global averages and errors of $g(r)$ in each bin.

In the Molecular Dynamics code <font face="Courier">MolDyn_NVE.cpp</font> the same changes have been applied to the function _MeasureEachStep_. The normalization is made by a loop in the _main_ and the global averages and errors are calculated together with those of the other properties in another loop.

The solid phase has a potential energy lower (higher in modulus) than the liquid phase and much lower than the gas phase. The opposite happens for the pressure, for which the solid phase has the highest value and the gas phase the lowest. This was also observed in the results of the MD simulations and reflects the fact that atoms in the solid phase are more strongly bound to each other than in the liquid phase, while at the gas density, which is extremely low, the attraction is so weak that even small temperatures give rise to a phase where atoms are free to move.

The radial distribution function curves from MD and MC simulations overlap almost perfectly. They are all null for small $r$, as the strongly repulsive term proportional to $r^{-12}$ in the Lennard-Jones potential prevents particles from getting too close to each other. They also have the highest peak near $r=a=1.09\sigma$ (vertical lines on the graphs), the theoretical lattice constant for a noble gas fcc crystal with a Lennard-Jones pairwise potential, namely the equilibrium ditance among nearest neighbours that minimizes the total potential energy. This is thus the distance at which two particles are most likely to be found. The plot for the solid phase has three more peaks, close to $r=\sqrt{2}a$, $r=\sqrt{3}a$ and $r=2a$. These values correspond to the distance between an atom and its second, third and fourth nearest neighbours in a perfect fcc lattice with lattice constant $a$. The first four correlation numbers of the fcc lattice (i.e. the number of atoms in the first four shells of neighbours) are: $12, 6, 24, 12$. The height of the peaks is not proportional to these numbers because of the normalization factor of $g(r)$, $(\rho N \Delta V(r))^{-1}$. Indeed

$$
\Delta V(r) =\frac{4\pi}{3} \left[ (r+dr)^3-r^3 \right] = 4\pi \left[ r^2 dr + r dr^2 +\frac{1}{3}dr^3 \right] \approx 4\pi r^2 dr
$$

where we have used $dr \ll r$, since $dr$ is the bin width and the first 40 bins of the histogram are empty. The height is then reduced proportionally to $r^2$ approximately. The following image shows the plot of $g(r)$ rescaled by $\Delta V(r)$: the peak heights are almost proportional to the correlation numbers (here scaled so that the fourth one perfectly matches the corresponding peak), which confirms that the structure of the system is compatible with that of a fcc lattice.

```python
import math

npart=108
rho=1.1
V=npart/rho
box=V**(1/3)

dv=np.loadtxt("deltav.out")

plt.figure(figsize=(15,7))

xsmd=(np.arange(len(g_solid_MD[:,0]))+1)*box/(2*nbins)*sigma
plt.errorbar(xsmd, g_solid_MD[:,0]*dv[:,1], yerr=g_solid_MD[:,1], label='MD NVE', color='green')
xsmc=(np.arange(len(g_solid_MC[:,0]))+1)*box/(2*nbins)*sigma
plt.errorbar(xsmc, g_solid_MC[:,0]*dv[:,1], yerr=g_solid_MC[:,1], label='MC NVT', color='orange')
dist=[1,math.sqrt(2),math.sqrt(3),2]
h=1.37727/12
corr_num=np.array([12,6,24,12])
nns=corr_num*h
for c in range(3):
    y=np.linspace(0, nns[c], 10)
    plt.plot([dist[c]*1.09*sigma for i in y], y, linewidth=8, color='dimgray')
c=3
y=np.linspace(0, nns[c], 10)
plt.plot([dist[c]*1.09*sigma for i in y], y, linewidth=8, color='dimgray', label='number of neighbours at a given distance in a perfect fcc lattice')
plt.grid(True)
plt.title('Non-normalized radial distribution function, solid')
plt.xlabel('$r(nm)$')
plt.ylabel('$g(r)x\Delta V(r)$')
plt.legend()
plt.show()
```

In the liquid phase $g(r)$ has a second peak before $r=2a$ but exhibits a less ordered structure than the solid phase. Any ordering disappears in the gas phase for $r>a$ and $g(r)$ becomes constant and equal to $1$, meaning that two particles are equally likely to be found at any distance and the 2-particles density is just $\rho_N^{(2)} (|\vec{r_1}-\vec{r_2}|) = \rho^2$, as for the ideal gas with a large number of particles $N$, because the presence of a particle in $\vec{r_1}$ doesn't affect the probability of having another particle in $\vec{r_2}$ for $|\vec{r_1}-\vec{r_2}|$ large enough (the joint probability is the product of two independent probabilities).
