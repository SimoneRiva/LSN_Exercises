During this exercise you will variationally optimize the ground state of a single quantum particle in a one dimensional (1D) space confined by the following external potential:
$$
V(x) = x^4 - \frac{5}{2}x^2
$$

This 1D quantum model is not analytically solvable and should be attacked by numerical methods. In order to obtain an approximate wave function for the ground state, the Variational Monte Carlo method uses a trial wave function, parametrized by a set of variational parameters. An approximate model for the ground state wave function is the following trial wave function:
$$
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}
$$

$\Psi_T^{\sigma,\mu}(x)$ depends on two variational parameters, $\sigma$, which is the width of the two Gaussians under square root, and $\mu$, which represents the distance from the origin of their average values.

### Exercise 08.1

Write a C++ Variational Monte Carlo code for a single quantum particle in 1D which exploits the Metropolis algorithm to sample a trial wave function $|\Psi_T(x)|^2$ using an uniform transition probability $T(x_{new}|x_{old})$ (Suggestion: as a starting point, you can use the code you have already developed for the Hydrogen atom or the Monte Carlo NVT code). Using data blocking, the code should be able to compute the expectation value for the Hamiltonian

$$
\langle {\hat H} \rangle_T = 
\frac{\int dx \Psi^*_T(x) {\hat H} \Psi_T(x)}
{\int dx |\Psi_T(x)|^2} = 
\int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2} \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

which can be used to find the parameters that minimize this quantity.

### Solution

The Variational Monte Carlo code <font face="Courier">VMC_1D.cpp</font> implements an optimization of the trial wavefunction's parameters through the simulated annealing (SA) algorithm in order to minimize the expectation value of the Hamiltonian $\langle {\hat H} \rangle_T$. The annealing schedule is made of $100$ cooling steps each with $n_i=100$, $i=1,...,100$, Monte Carlo steps.

Given a starting guess for the values of the two parameters $\mu$ and $\sigma$, at each Monte Carlo step the function _SAStep_ suggests a possible move by increasing $\mu$ and $\sigma$ by a uniformely distributed (in $(-0.5,0.5)$ and $(-0.2,0.2)$ respectively) random number and evaluates the average of

$$
E_{loc}(x) = \frac{{\hat H} \Psi_T(x)}{\Psi_T(x)} = \frac{\left( -\frac{1}{2}\frac{d^2}{dx^2} + V(x) \right) \Psi_T(x)} {\Psi_T(x)} \\
= -\frac{1}{2} \left[ \left( \frac{(x-\mu)^2}{\sigma^4}-\frac{1}{\sigma^2} \right) e^{-\frac{(x-\mu)^2}{2\sigma^2}} + \left( \frac{(x+\mu)^2}{\sigma^4}-\frac{1}{\sigma^2} \right) e^{-\frac{(x+\mu)^2}{2\sigma^2}} \right] \frac{1}{\Psi_T(x)} + V(x)
$$

on $10^5$ points $x$ sampled from the distribution

$$
p(x) = \frac{|\Psi_T(x)|^2}{\int dx' |\Psi_T(x')|^2} 
$$

using the Metropolis algorithm. Actually $\sigma$ is not allowed to become lower than $0.1$, because for small $\sigma$ the code fails to return $E_{loc}(x)$, which contains terms proportional to $\sigma^{-4}$, and returns _nan_ instead. The move is then accepted with probability

$$
\alpha = min(1, e^{-\beta(E_{new}-E)})
$$

where $E_{new}$ is the average energy (mean of the Hamiltonian on the trial wavefunction) calculated with the new guess for the parameters, while $E$ is the average energy at the previous step.

Evaluating the blocking average of $E_{loc}(x)$ is the task of the routine _TrialEnergy_, which equilibrates the Metropolis sampling and then computes only its final average with no error, as it is only necessary for  comparison with the previous energy (namely to be substituted to $E_{new}$ in the expression above). $N=100$ blocks of $L=10^3$ steps each are used for this purpose. _BestEnergy_ computes and prints on the output file <font face="Courier">energy.out</font> all the blocking averages and their uncertainties once the optimal parameters have been found and also prints all the sampled points on <font face="Courier">points.out</font>.

Both these routines call the function _MetroStepUnif_, that works exactly as its namesake class in exercise 05, sampling a generic probability (_prob_ $=|\Psi_T(x)|^2$ in this case) which is passed to it through a pointer to an abstract class _BaseFunction_. One thing one has to pay attention to before calling _TrialEnergy_ or _BestEnergy_ is to always set the current values of $\mu$ and $\sigma$ in the class _prob_ with the specific method (_prob->Set(mu, sigma)_). Indeed they are necessary not only for the evaluation of $E_{loc}(x)$, which uses the global variables that have already been moved by the SA, but also for the sampling of $p(x)$, which is calculated using the data members of _prob_. 

All the accepted moves of the parameters during the SA search are recorded on <font face="Courier">energy_search.out</font> on four columns containing the values of $\beta$, $\mu$, $\sigma$ and the expectation value of the Hamiltonian on the trial wavefunction with those parameters, as a check.

### Solution

After several attempts the annealing schedule was set with inverse temperatures $\beta_i = 5i$, $i=1,...,100$. This way the starting temperature is high enough to explore the space of possible parameters without being trapped in a local well and the final temperature is low enough to have minimal oscillations away from the (hopefully) global minimum, as one can see from <font face="Courier">energy_search.out</font>.

The simulation returned the following best values: $\mu=-0.818203$, $\sigma=0.613058$. Here the sign of $\mu$ is obviously irrelevant, as $\Psi_T(x)$ is invariant for its change. The variational ground state energy is $-0.445247$. Units are chosen so that $\hbar=1$ and $m=1$.

The graph below reports the average value of $\langle {\hat H} \rangle_T$ with its uncertainty as a function of the number of blocks with the optimized parameters.

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

HT = np.loadtxt("VMC_1D/energy.out")
x=np.arange(len(HT[:,0]))
plt.figure(figsize=(15,8))
plt.errorbar(x, HT[:,0], yerr=HT[:,1], color='royalblue', label='block average')
plt.plot(x, [-0.46046587969271535 for i in x], color='midnightblue', label='exact ground state')
plt.title('Expectation value of the Hamiltonian on $\psi_T$ with optimized parameters')
plt.xlabel("# blocks")
plt.ylabel("$<H>_T$")
plt.legend()
plt.grid(True)
plt.show()
```

The variational energy is close to the ground state energy even if not compatible. We could have expected this though, since we have chosen a specific parametric form for the trial wavefunction (the sum of two gaussians), which in general will not represent the real form of the ground state wavefunction, so it will be impossible to approximate the ground state arbitrarily well.

The following image shows the distribution of points sampled with Metropolis for the optimized parameters, which turns out to be similar to the numerical solution for $|\Psi_{GS}(x)|^2$ obtained from the matrix equation.

```python
y = np.loadtxt("VMC_1D/points.out")

def Vpot(x):
    return (x**2 - 2.5)*x**2

hbar = 1
m = 1
a = 10
N = 1000

x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0]
V = Vpot(x)

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

E,psi = np.linalg.eigh(H)

psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

plt.figure(figsize=(15,8))
scale = 0.3
plt.plot(x, scale*V, color="Black", label="potential")
plt.hist(y, density=True, bins=100, color='gold', label='sampled')
plt.plot(x,(psi[0])**2, color='orange', label='matrix equation solution')
plt.title('Potential & $|\Psi_T(x)|^2$')
plt.xlabel("x")
plt.grid(True)
plt.legend()
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))
plt.show()
```

### Supplementary material

How can we describe a wave function as a vector? We set up a lattice of discrete points, spaced by a constant distance $dx$, and record the value of the function at each lattice point. In this way the wave function, $\Psi(x)$ become a vector:

$$
\Psi(x) \to \left[ \Psi(x_1), \Psi(x_2), ..., \Psi(x_N) \right] = 
\left[ \Psi_1, \Psi_2, ..., \Psi_N \right]
$$

The time independent Schrodinger equation 

$$
{\hat H} \Psi(x) = \left( -\frac{\hbar^2}{2m}\frac{\partial^2}{\partial x^2} + V(x) \right) \Psi(x) = E \Psi(x)
$$

can be written in matrix form. But how do we write the second derivative at a particular point? Let's start with the first derivatives:
$$
\frac{\partial}{\partial x} \Psi\left(x=(x_i+x_{i+1})/2\right) \simeq \frac{\Psi_{i+1}-\Psi_{i}}{dx}\\
\frac{\partial}{\partial x} \Psi\left(x=(x_i+x_{i-1})/2\right) \simeq \frac{\Psi_{i}-\Psi_{i-1}}{dx}
$$

Thus, we can discretize the second derivative in the following way:
$$
\frac{\partial^2}{\partial x^2} \Psi(x=x_i) =
\frac{\frac{\partial}{\partial x} \Psi\left(x=(x_i+x_{i+1})/2\right) - \frac{\partial}{\partial x} \Psi\left(x=(x_i+x_{i-1})/2\right)}{dx} =
\frac{\Psi_{i+1}-2\Psi_{i}+\Psi_{i-1}}{dx^2}
$$

We finally obtain:

$$
\begin{bmatrix}
 \frac{\hbar^2}{m\,dx^2}+V(x_1)& -\frac{\hbar^2}{2m\,dx^2} & 0 & \dots  & 0 \\
-\frac{\hbar^2}{2m\,dx^2} & \frac{\hbar^2}{m\,dx^2}+V(x_2) & -\frac{\hbar^2}{2m\,dx^2} & \dots  & 0 \\
 0                        & -\frac{\hbar^2}{2m\,dx^2} & \frac{\hbar^2}{m\,dx^2}+V(x_3) & \dots  & 0 \\
 \vdots                   & \vdots                    & \vdots & \ddots & \vdots \\
 0                    & 0                & 0 & \dots  & \frac{\hbar^2}{m\,dx^2}+V(x_N)
\end{bmatrix}
\begin{bmatrix}
\Psi_{1} \\
\Psi_{2} \\
\vdots \\
\Psi_{N}
\end{bmatrix}
= E
\begin{bmatrix}
\Psi_{1} \\
\Psi_{2} \\
\vdots \\
\Psi_{N}
\end{bmatrix}
$$

To solve the time independent Schrodinger equation we therefore have to find eigenstates and eigenvalues of the matrix just obtained.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = np.linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = np.diag(np.ones(N-1),-1)-2*np.diag(np.ones(N),0)+np.diag(np.ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + np.diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = np.linalg.eigh(H)

# Take the transpose & normalize
psi = np.transpose(psi)
psi = psi/np.sqrt(dx)

print("Ground state energy: ", E[0])
print("1st excited state energy: ", E[1])
print("2nd excited state energy: ", E[2])

# Plot a few things
plt.figure(figsize=(8,5))
scale = 0.3
plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.plot(x,(psi[0])**2)
plt.plot(x,(psi[1])**2)
plt.plot(x,(psi[2])**2)
plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))
plt.show()
```
