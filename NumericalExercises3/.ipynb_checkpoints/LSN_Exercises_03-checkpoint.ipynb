### Exercise 03.1 <span style="color:red"> Plain vanilla option pricing </span>

Black-Scholes theory assumes that the time evolution of the asset price is stochastic and exhibits geometric Brownian motion (GBM) with constant risk-free interest rate $r$ and volatility $\sigma$.

Black-Scholes <span style="color:blue"> analytic solution </span> for the price at time t<T:
- European Call-option price:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- European Put-option price:
$$ C\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

where 
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

where $d_2 = d_1 - \sigma\sqrt{T-t}$, and where
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
is the cumulative of Gaussian $\sim \mathcal{N}(\mu=0,\sigma^2=1)$,
and where
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$

By using the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

compute at time $t=0$ **via Monte Carlo** the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$:

1. <span style="color:red"> By sampling **directly** the final asset price $S(T)$</span> for a $GBM(r,\sigma^2)$
2. <span style="color:red"> By sampling the discretized $GBM(r,\sigma^2)$ path of the asset price</span> 
dividing $[0,T]$ in $10^2$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$

<p style="border:2px; border-style:solid; border-color:#F5F5F5; padding: 1em; background-color:#F5F5F5">
    Remember that given an asset price $S\sim GBM(\mu,\sigma^2)$, with initial value $S(0)$, then
    $$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$
    where $W(t)\sim\mathcal{N}(0,t)$.
    Moreover, since the increments of $W$ are independent and normally distributed, the simple recursive procedure for sampling discretized values of $S\sim GBM(\mu,\sigma^2)$ at $t_1 < t_2 < â€¦ < t_n$ is:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    with $Z_i \sim \mathcal{N}(0,1)$
</p>

<span style="color:red">Make four pictures for the estimation of the European call-option prices, $C[S(0),0]$ (direct and discretized), and put-option prices, $P[S(0),0]$ (direct and discretized), with their uncertainties </span> with a large number of asset prices at time $t=T$, say $M\ge 10^4$, as a function of the number of blocks, $N$</font>. <span style="color:blue"> As usual, in computing statistical uncertainties, use data blocking.</span>

### For comparison, Black-Scholes analytic solution:

```python
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

### Solution

#### Direct sampling

The code produces $M=10^4$ simulatons of the final price of the asset
$$S(T) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\right]$$
by drawing $W(T)$ from a gaussian distribution with mean $\mu=0$ and variance $\sigma^2=T$. The function _DataBlocking_ then divides the simulations into $N=100$ blocks and computes the blocking averages of two different functions of $S(T)$, namely the price of a call and a put european option:
$$CallPrice [S(T),T] = e^{-rT}max(0,S(T)-K) \\
PutPrice [S(T),T] = e^{-rT}max(0,K-S(T))$$

The results for the price of both options are stable and compatible with the analytic results of Black-Scholes after a few tens of blocks.

#### Discrete path sampling

Here the geometric brownian motion is built step by step by means of
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
where $Z_i$, $i=1,...,N_{steps}$ with $N_{steps}=100$, are obtained from a normal gaussian distribution. After generating the whole path and obtaining the final value $S(T)$ for all the $M=10^4$ simulations, these are processed as in the previous case.

The results agree with the analytic solutions again and the errors are comparable with those obtained from direct sampling.

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

call_direct=np.loadtxt("call_direct_sampling.out")
call_discrete=np.loadtxt("call_discrete_sampling.out")
put_direct=np.loadtxt("put_direct_sampling.out")
put_discrete=np.loadtxt("put_discrete_sampling.out")

fig, ax = plt.subplots(2, 2, figsize=(15, 15))

ax[0,0].errorbar(call_direct[:,0], call_direct[:,1], yerr=call_direct[:,2], label='simulations')
ax[0,0].plot(call_direct[:,0], [call for x in call_direct[:,0]], color='midnightblue', label='analytic result')
ax[0,0].legend()
ax[0,0].set_title('Call option with direct calculation');

ax[0,1].errorbar(put_direct[:,0], put_direct[:,1], yerr=put_direct[:,2], label='simulations')
ax[0,1].plot(put_direct[:,0], [put for x in put_direct[:,0]], color='midnightblue', label='analytic result')
ax[0,1].legend()
ax[0,1].set_title('Put option with direct calculation');

ax[1,0].errorbar(call_discrete[:,0], call_discrete[:,1], yerr=call_discrete[:,2], label='simulations')
ax[1,0].plot(call_discrete[:,0], [call for x in call_discrete[:,0]], color='midnightblue', label='analytic result')
ax[1,0].legend()
ax[1,0].set_title('Call option with discrete calculation');

ax[1,1].errorbar(put_discrete[:,0], put_discrete[:,1], yerr=put_discrete[:,2], label='simulations')
ax[1,1].plot(put_discrete[:,0], [put for x in put_discrete[:,0]], color='midnightblue', label='analytic result')
ax[1,1].legend()
ax[1,1].set_title('Put option with discrete calculation');

plt.setp(ax[:, :], xlabel='# blocks')
plt.setp(ax[:, :], ylabel='price')

for row in ax: 
    for x in row: 
        x.grid(b=True)
```
