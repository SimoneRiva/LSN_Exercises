```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np


with open("direct_sampling.txt", "r") as inp:
    read = inp.readlines()
    data=[line.split() for line in read]
    Tdata = list(zip(*data))
    for i, _ in enumerate(Tdata):
        Tdata[i]=list(Tdata[i])
        for j, _ in enumerate(Tdata[i]):
            Tdata[i][j]=float(Tdata[i][j])

with open("discrete_sampling.txt", "r") as inp2:
    read2 = inp2.readlines()
    data2=[line.split() for line in read2]
    Tdata2 = list(zip(*data2))
    for i, _ in enumerate(Tdata2):
        Tdata2[i]=list(Tdata2[i])
        for j, _ in enumerate(Tdata2[i]):
            Tdata2[i][j]=float(Tdata2[i][j])

            
fig, ax = plt.subplots(2, 2, figsize=(15, 15))

ax[0,0].errorbar(Tdata[0],Tdata[1],yerr=Tdata[2], label='simulations')
ax[0,0].plot(Tdata[0], [14.975790778311286 for x in Tdata[0]], color='midnightblue', label='analytic result')
ax[0,0].legend()
ax[0,0].set_title('call option with direct caclulation');
ax[0,1].errorbar(Tdata[0],Tdata[3],yerr=Tdata[4], label='simulations')
ax[0,1].plot(Tdata[0], [5.4595325819072364 for x in Tdata[0]], color='midnightblue', label='analytic result')
ax[0,1].legend()
ax[0,1].set_title('put option with direct caclulation');
ax[1,0].errorbar(Tdata2[0],Tdata2[1],yerr=Tdata2[2], label='simulations')
ax[1,0].plot(Tdata2[0], [14.975790778311286 for x in Tdata2[0]], color='midnightblue', label='analytic result')
ax[1,0].legend()
ax[1,0].set_title('call option with discrete caclulation');
ax[1,1].errorbar(Tdata2[0],Tdata2[3],yerr=Tdata2[4], label='simulations')
ax[1,1].plot(Tdata2[0], [5.4595325819072364 for x in Tdata2[0]], color='midnightblue', label='analytic result')
ax[1,1].legend()
ax[1,1].set_title('put option with discrete caclulation');
plt.setp(ax[:, :], xlabel='#simulations')
plt.setp(ax[:, :], ylabel='price')

for row in ax: 
    for x in row: 
        x.grid(b=True)
```

### For comparison, Black-Scholes analytic solution:

```python
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```
